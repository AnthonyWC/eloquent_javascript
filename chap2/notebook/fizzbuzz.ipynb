```javascript
function fizzbuzz(n) {
    var i = 1;
    
    function divide(k, div, output) {
        if (k % div === 0) {
           return true
        } else {
           return false 
        }
    }
    
    while (i <= n) {
        if (divide(i, 3, "Fizz") && divide(i, 5, "Buzz")) {
            console.log("FizzBuzz");    
        } else if (divide(i, 3, "Fizz")) {
            console.log("Fizz"); 
        } else if (divide(i, 5, "Buzz")) {
            console.log("Buzz")    
        } else {
            console.log(i)
        }
        i += 1;
    }
}

fizzbuzz(50)
```
